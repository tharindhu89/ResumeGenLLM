```python
import pandas as pd
```

```python
df = pd.read_csv('similarity.csv')
```

```python
df.head()
```

```python
df_cleaned = df.dropna()
```

```python
df_cleaned.head()
```

```python
# Select the relevant columns
similarity_columns = ['Similarity: Original vs Similar', 'Similarity: Original vs Optimized', 'Similarity: Similar vs Optimized']

# Calculate basic statistics
statistics = df[similarity_columns].describe()

# Display the basic statistics
print(statistics)

# Calculate the correlation matrix between the three similarity columns
correlation_matrix = df[similarity_columns].corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for each similarity column
plt.figure(figsize=(12, 6))
df[similarity_columns].hist(bins=20, edgecolor='black', figsize=(12, 6))
plt.suptitle('Distribution of Similarity Scores', fontsize=16)
plt.show()

# Plot a correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Similarity Scores', fontsize=16)
plt.show()
```

```python
sns.pairplot(df[similarity_columns])
plt.suptitle('Pairwise Scatter Plots of Similarity Measures', fontsize=16)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = df[similarity_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Heatmap of Similarity Scores', fontsize=16)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df[similarity_columns])

# Create a DataFrame to store PCA results
pca_df = pd.DataFrame(pca_result, columns=['PCA 1', 'PCA 2'])

# Plot the PCA result
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PCA 1'], pca_df['PCA 2'], c='blue', edgecolor='black')
plt.title('PCA of Similarity Measures', fontsize=16)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Bivariate regression between 'Similarity: Original vs Optimized' and 'Similarity: Similar vs Optimized'
X = df[['Similarity: Original vs Optimized']].values
y = df['Similarity: Similar vs Optimized'].values

model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

# Plot the regression
plt.figure(figsize=(10, 6))
plt.scatter(df['Similarity: Original vs Optimized'], df['Similarity: Similar vs Optimized'], color='blue')
plt.plot(df['Similarity: Original vs Optimized'], predictions, color='red', linewidth=2)
plt.title('Regression: Original vs Optimized vs Similar vs Optimized', fontsize=16)
plt.xlabel('Similarity: Original vs Optimized')
plt.ylabel('Similarity: Similar vs Optimized')
plt.show()

# Calculate the MSE
mse = mean_squared_error(y, predictions)
print(f"Mean Squared Error: {mse}")
```

```python
from sklearn.cluster import KMeans

# Apply K-means clustering with 3 clusters (can adjust as needed)
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[similarity_columns])

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Similarity: Original vs Optimized', y='Similarity: Similar vs Optimized', 
                hue='Cluster', data=df, palette='Set1', s=100)
plt.title('K-Means Clustering of Similarity Measures', fontsize=16)
plt.xlabel('Similarity: Original vs Optimized')
plt.ylabel('Similarity: Similar vs Optimized')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[similarity_columns], palette='Set2')
plt.title('Box Plots of Similarity Measures', fontsize=16)
plt.ylabel('Similarity Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Similarity: Original vs Similar'], shade=True, label='Original vs Similar')
sns.kdeplot(df['Similarity: Original vs Optimized'], shade=True, label='Original vs Optimized')
sns.kdeplot(df['Similarity: Similar vs Optimized'], shade=True, label='Similar vs Optimized')
plt.title('Density Plot of Similarity Scores', fontsize=16)
plt.xlabel('Similarity Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```
